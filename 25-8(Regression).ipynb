2. Data Cleaning

```python
# Handling Missing Values
df.isnull().sum()
```

```python
# Fill missing values
df['price'].fillna(df['price'].mean(), inplace=True) # inplace - does changes in the table itself
df
```

```python
#Check for duplicate values
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace = True)
```
