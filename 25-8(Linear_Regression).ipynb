# Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
```

```python
#Linear rel btw hrs of study n score
hours_studied=np.random.uniform(0,10,100) #Record for 100 students between 1 to 10
exam_scores=50+5 *hours_studied+np.random.normal(0,2,100)  #.normal is for error
```

```python
data_df=pd.DataFrame({
    'Hours_studies':hours_studied ,
    'Exam_score':exam_scores
})
data_df
```

```python
#Data preprocessing
```

# Train, Test Split

```python
x=data_df[['Hours_studies']] #independent variable
y=data_df[['Exam_score']] #depedent variable
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2, random_state=42)# random state= same state
#train_testx_test,split is function which will return 4 parameters
#x_train-training data for x
#y_train-training data for y
#x_test-testing data for x
#y_test-testing data for y
```

```python
x_train.shape,x_test.shape
```

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
#Use a object of model to create 
model=LinearRegression()
model.fit(x_train,y_train)
```

# Model Evaluation

```python
# Prediction
y_pred=model.predict(x_test)
#create a prediction variable and use fucntion predict to pass independent variable 
```

# Visualization

```python
plt.scatter(x_test,y_test,color='blue', label='Actual Data')
#plt.plot(x_test,y_test,color='blue', label='Actual Data')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')

#plt.plot is used to plot the line
plt.plot(x_test,y_pred,color='Red',label='Regression', linewidth=2 )

#Title
plt.title('Linear Regression Line')
#Display that label in the 
plt.legend()
#show the plot
plt.show()
```
